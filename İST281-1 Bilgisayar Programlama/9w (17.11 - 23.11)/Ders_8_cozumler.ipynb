```python
# Soru 2 
def ciftek(n):
    sonuc = 0
    while sonuc != 1 :
        if n%2==0 : 
            print(n//2)
            sonuc = n//2
            n = n//2
        else :
            print(3*n+1)
            sonuc = 3*n+1
            n = 3*n+1

ciftek(6)
ciftek(7)
```

```python
# Soru 3

def perfect(n) :
    sonuc = 0
    bolenler=[]
    for i in range(1,n) :
        if n%i == 0 :
            bolenler.append(i)
    for j in bolenler :
        sonuc += j
    if sonuc is n : 
        print(True)
    else :
        print(False)
        while sonuc != n :
            n += 1
            bolenler=[]
            sonuc = 0
            for i in range(1,n) :
                if n%i == 0 :
                    bolenler.append(i)
            for j in bolenler :
                sonuc += j
            if sonuc == n :
                print(True)
                print(n)
                break

perfect(5)
perfect(6)
```

```python
# Soru 4 

liste = [4,5,4,4,6,5]
yeni_liste = []
def temiz(L) :
    """Listede tekrar edenleri temizleyen bir fonksiyon"""
    for i in L:
        if i not in yeni_liste:
            yeni_liste.append(i)
    print(yeni_liste)

temiz(liste)
```

```python
# Soru 5
def taban_alani(r):
    """Silindirin taban alanini hesaplar"""
    return 3.14*(r**2)
    
def yanal_alan(r,h): 
    """Silindirin yanal alanini hesaplar"""
    return 2*3.14*r*h

def toplam_alan(r,h):
    """Silindirim toplam alanini hesaplar"""
    return taban_alani(r) + yanal_alan(r,h)

print(taban_alani(5))
print(yanal_alan(5,10))
print(toplam_alan(5,10))
```
